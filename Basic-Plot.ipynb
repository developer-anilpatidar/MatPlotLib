# Legend

```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,54,55,51,52,47,49]
min_t = [43,35,38,39,42,32,30]
avg_t = [48,45,40,46,43,44,49]
```

```python
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

Add Label and Legend

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

plt.legend()
```

We can also add style for Legend

```python
# loc - Location of the legend like upper right, upper left, lower right, lower center etc
# fontzise - fontsize of the legend text
# shadow - for legend box shadow
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')
plt.legend(loc='lower left', fontsize=12, shadow=True)
```

```python
# loc - use best for automatic adjustment
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')
plt.legend(loc='best', fontsize=12)
```

Using bbox_to_anchor for precise placement:
This method provides more granular control over the legend's position, especially for placing it outside the plot area or at a specific coordinate.

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')
plt.legend(loc='best', bbox_to_anchor=(0.5, -0.15))
```

# Grid

Add Grid using Grid() function

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')
plt.legend(loc='best', fontsize='small')

plt.grid()
```
