```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]     # days
y = [30,35,49,31,46,52,54]      #Temprature      
```

```python
plt.plot(x, y)
```

Use Styles with Plot

```python
plt.plot(x, y, color='green', linestyle='dotted')
```

Give Lable to Axis and Title of Plot

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color='orange', linestyle='dashed')
```

```python
plt.plot(x,y,'g')
```

```python
plt.plot(x,y, 'g*')
```

```python
plt.plot(x,y, 'r*--')
```

Marker Style

```python
plt.plot(x,y, marker='+', markersize=20)
```

Use alpha to make the transparent, it's scale between 0 - 1

```python
plt.plot(x,y, alpha=0.1)
```

```python
plt.plot(x,y, alpha=0.5)
```
