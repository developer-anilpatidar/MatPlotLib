ðŸ”¹ 1. Basic Vertical Bar Chart

```python
import matplotlib.pyplot as plt
```

```python
products = ['Pen', 'Pencil', 'Notbook', 'Eraser']
sales = [150, 120, 50, 90]
```

```python
plt.bar(products, sales)

plt.xlabel("Products")
plt.ylabel("Units Sold")
plt.title("Product Sales by Store")
plt.show() # show() is not nessesary but It's clearer and gives you more control
```

---

ðŸ”¹ 2. Change Bar Color & Width

```python
plt.bar(products, sales, color='orange', width=0.4)
```

âœ… Customize color and bar width.

---

ðŸ”¹ 3. Horizontal Bar Chart using barh

```python
plt.barh(products, sales)
```

âœ… Switch X and Y axis â€“ useful when category names are long.

___

ðŸ”¹ 4. Add Value Labels on Bars

```python
bars = plt.bar(products, sales, color='green')

# Annotate each bar with value
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, yval, ha='center')

plt.title('Product Sales with Labels')
plt.show()
```

âœ… Shows data values on top of each bar.

ðŸ”¹ 5. Compare Two Categories (Grouped Bar Chart)

```python
import numpy as np
```

```python
products = ['Pen', 'Pencil', 'Notebook', 'Eraser']
store1 = [150, 120, 90, 60]
store2 = [130, 100, 70, 80]
```

```python
x = np.arange(len(products))
```

```python
x
```

```python
width = 0.35
```

```python
plt.bar(x - width/2, store1, label='Store A', color='blue')
```

```python
plt.xticks(x, products) # Use eticks to replace value with products
plt.bar(x - width/2, store1, label='Store A', color='blue')
```

```python
plt.xticks(x, products)
plt.bar(x - width/2, store1, width, label='Store A', color='blue')
plt.bar(x + width/2, store2, width, label='Store B', color='orange')
plt.show()
```

ðŸ“Œ Show with Lables and Legend

```python
plt.xticks(x, products)
plt.xlabel('Products')
plt.ylabel('Units Sold')
plt.title('Product Sales by Store')

plt.bar(x - width/2, store1, width, label='Store A', color='blue')
plt.bar(x + width/2, store2, width, label='Store B', color='orange')

plt.legend()
plt.show()
```

âœ… Compare sales between two stores side-by-side.

---

ðŸ”¹ 6. Customize Grid & Style

```python
plt.xticks(x, products)
plt.xlabel('Products')
plt.ylabel('Units Sold')
plt.title('Product Sales by Store')

plt.bar(x - width/2, store1, width, label='Store A', color='blue')
plt.bar(x + width/2, store2, width, label='Store B', color='orange')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.legend()
plt.show()
```
