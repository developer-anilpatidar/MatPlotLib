# ðŸ“Š What is a Histogram?

A Histogram is used to visualize the distribution of a dataset.

It divides the data into intervals (bins) and shows the frequency (count) of values in each bin.

ðŸŸ¦ When to Use a Histogram

âœ… Use a Histogram When:

You want to understand the distribution of numeric data

â†’ e.g., how are ages, incomes, scores, or weights spread out?

You want to group continuous data into intervals (bins)

â†’ e.g., how many people fall between ages 20â€“30, 30â€“40, etc.


```python
import matplotlib.pyplot as plt
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 77, 80, 83, 130]
```

```python
plt.hist(blood_sugar)
```

ðŸ“Œ By Default it pins 10 bins

```python
plt.hist(blood_sugar, bins=3)
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.90)
```

ðŸ“Œ We can also customise range using bin

```python
plt.hist(blood_sugar, bins=[80,100,120,150], rwidth=0.95, color='g')
```

# With 2 size of data sets

```python
blood_sugar_man = [113, 85, 90, 150, 149, 88, 93, 77, 80, 83, 130]
blood_sugar_woman = [69, 95, 70, 109, 90, 139, 108, 145, 60, 120, 135]
```

```python
plt.hist([blood_sugar_man, blood_sugar_woman], bins=[80,100,120,150], rwidth=0.95)
```

ðŸ“Œ We can add colors

```python
plt.hist([blood_sugar_man, blood_sugar_woman], bins=[80,100,120,150], rwidth=0.95, color=['green', 'orange'])
```

ðŸ“Œ We can add Labels

```python
plt.xlabel("Sugar Range")
plt.ylabel("Total no of patients")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_man, blood_sugar_woman], bins=[80,100,120,150], rwidth=0.95, color=['green', 'orange'], label=['Man', 'Woman'])
plt.legend()
```

ðŸ“Œ We can also show in Horizontal

```python
plt.xlabel("Sugar Range")
plt.ylabel("Total no of patients")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_man, blood_sugar_woman], bins=[80,100,120,150], rwidth=0.95, color=['green', 'orange'], label=['Man', 'Woman'],
         orientation='horizontal')
plt.legend()
```
